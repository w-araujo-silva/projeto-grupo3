```python
import pandas as pd


# lendo o dataset
def carregar_banco_csv(file_path='/content/data_science_salaries.csv'):
    return pd.read_csv(file_path)
```

```python
df = carregar_banco_csv()
```

```python
df
```

```python
def media_salario(df):
    return round(df["salary_in_usd"].mean(), 2)
```

## Média Salarial em USD

```python
print(media_salario(df))
```

## Conclusões
A média salarial no periodo foi de USD 145.560,56.

## Evolução temporal
Como os salários médios variaram de 2020 a 2024?

```python
cols = ['salary_in_usd']
cols2 = ['work_year']
df.groupby(cols2)[cols].mean()
```

## Conclusões
Houve uma redução no salário médio em 2021 em relação ao ano anterior, porém nos anos seguintes pode ser constatar uma crescente na média salarial.

## Evolução temporal
Média salarial por ano e modalidade de trabalho

```python
cols = ['salary_in_usd']
cols2 = ['work_models', 'work_year']
df.groupby(cols2)[cols].mean()
```

## Conclusões
Nas modalidades de trabalho presencial e remota é possível perceber uma crescente na média salárial, exceto no ano de 2021 na modalidade remota, onde houve uma redução em comparação ao ano anterior. Comparando ano a ano as modalidades presencial e remota, a modalidade presencial só possui média salarial inferior a remota nos anos de 2020 e 2021, possivelmente por haver mais funcionários na modalidade remota em decorrencia da pandemia de COVID 19, pois lidera nos demais anos. A modalidade hibrida apresentou oscilação nas médias salariais no decorrer dos anos, havendo um crescimento expressivo no ano de 2024.

## Evolução temporal
Houve mudanças significativas na modalidade de trabalho (mais remoto, menos remoto) no período?

```python
cols = ['salary_in_usd']
cols2 = ['work_models', 'work_year']
df.groupby(cols2)[cols].count()
```

## Conclusões
Considerando o intervalo de 2020 até 2023 e desconsiderando o ano de 2024, pois possívelmente apresenta dados parciais, é possível perceber uma crescente na quantidade de funcionários nas modalidades de trabalho presencial e remoto. Nos anos de 2020 a 2022 a modalidade remota foi a mais praticada, possivelmente por conta da pandemia de COVID 19, porém em 2023 o trabalho presencial liderou como modalidade mais utilizada. A modalidade de trabalho hibrida apresenta a menor quantidade de adeptos em todos os anos quando comparada as demais modalidades.
